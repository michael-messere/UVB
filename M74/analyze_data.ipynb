```python
def wavelength_to_energy(wavelength_nm):
    # Constants
    h = 6.62607015e-34  # Planck's constant in J·s
    c = 299792458  # Speed of light in m/s
    
    # Convert wavelength to meters
    wavelength_m = wavelength_nm * 1e-9  # Convert from nm to meters
    
    # Calculate energy in joules
    energy_joules = h * c / wavelength_m
    
    # Convert energy from joules to ergs
    energy_erg = energy_joules * 1e7  # 1 Joule = 10^7 ergs
    
    return energy_erg

# Example usage:
wavelength_nm = 656.281  # Wavelength in nanometers
energy_erg = wavelength_to_energy(wavelength_nm)
print("Energy:", energy_erg, "ergs")
```
